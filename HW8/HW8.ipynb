```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
beta_values = np.linspace(0, 4, 1500)
theta = np.random.uniform(0, 2 * np.pi, len(beta_values))

beta_list = []
theta_list = []

for i in range(1000):
    theta = (theta + beta_values * np.sin(theta)) % (2 * np.pi)
    if i >= 800:
        beta_list.append(beta_values)
        theta_list.append(theta)

plt.figure(figsize=(10, 6))
plt.plot(beta_list, theta_list, ',k', alpha=0.1)
plt.xlabel(r'$\beta$')
plt.ylabel(r'$\theta$')
plt.title(r'Bifurcation Diagram ($\alpha = 0$)')
plt.xlim(0, 4)
plt.ylim(0, 2 * np.pi)
plt.grid(True, alpha=0.3)
plt.show()
```
